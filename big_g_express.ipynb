```python
import pandas as pd
from geopy.distance import geodesic
```

```python
lat1 = 36.0666667
lon1 = -86.4347222
lat_long = on_faults[['Latitude', 'Longitude']]

point1 = (lat1, lon1)
distances = []

for index, row in lat_long.iterrows():
    point2 = (row['Latitude'], row['Longitude'])
    distance = geodesic(point1, point2).mi
    distances.append(distance)

on_faults['Distance_A'] = distances
```
