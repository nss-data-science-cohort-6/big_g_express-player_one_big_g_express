```python
import pandas as pd
from geopy.distance import geodesic
```

```python
on_faults = pd.read_csv('data/J1939Faults.csv')
diagnostic = pd.read_csv('data/VehicleDiagnosticOnboardData.csv')
fault_code = pd.read_excel('data/Service Fault Codes_1_0_0_167.xlsx')
```

```python
lat1 = 36.0666667
lon1 = -86.4347222
lat_long = on_faults[['Latitude', 'Longitude']]

point1 = (lat1, lon1)
distances = []

for index, row in lat_long.iterrows():
    point2 = (row['Latitude'], row['Longitude'])
    distance = geodesic(point1, point2).mi
    distances.append(distance)

on_faults['Distance_A'] = distances
```
